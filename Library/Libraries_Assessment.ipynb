## 5. Convert Order Date and Ship Date to datetime

```python
date_cols = ['Order Date', 'Ship Date']
for col in date_cols:
    retail[col] = pd.to_datetime(retail[col], errors='coerce')
    print(f"{col} NaNs: {retail[col].isna().sum()}")

# 2. Clean and Convert Price Columns
price_cols = ['Cost Price', 'Retail Price', 'Profit Margin', 'Sub Total',
              'Discount $', 'Order Total', 'Shipping Cost', 'Total']

for col in price_cols:
    # Remove currency symbols and commas, then convert to numbers
    retail[col] = retail[col].replace(r'[\$,]', '', regex=True)
    retail[col] = pd.to_numeric(retail[col], errors='coerce')
    print(f"{col} NaNs: {retail[col].isna().sum()}")

print("\nFinal Data Types:")
print(retail[date_cols + price_cols].dtypes)
```

## 6. Check for future or inconsistent shipping dates


```python
inconsistent_shipments = retail[retail['Ship Date'] < retail['Order Date']]
if inconsistent_shipments.empty:
    print("No inconsistent shipping dates found (all Ship Dates are on or after Order Dates).")
else:
    print("Found inconsistent shipping dates:")
    print(inconsistent_shipments[['Order No', 'Order Date', 'Ship Date']])
```

## 7. Convert price columns to numeric (remove $ and commas).

```python
    # 1. Clean characters ($, and ,) from all price columns at once
retail[price_cols] = retail[price_cols].replace(r'[\$,]', '', regex=True)

# 2. Convert all price columns to numeric (NaN for errors)
retail[price_cols] = retail[price_cols].apply(pd.to_numeric, errors='coerce')

# 3. Quickly check results
print("Number of corrupted (NaN) entries per column:")
print(retail[price_cols].isna().sum())

print("\nNew Data Types:")
print(retail[price_cols].dtypes)
```

## 14. Are all discount percentages matching discount dollar amounts?


```python
# Convert to numeric
cols = ['Discount %', 'Discount $', 'Retail Price', 'Order Quantity']
for col in cols:
    retail[col] = pd.to_numeric(retail[col], errors='coerce')

# Remove missing values
retail = retail.dropna(subset=cols)

# Calculate difference
retail['Discount Diff'] = (retail['Retail Price'] * retail['Order Quantity'] * (retail['Discount %'] / 100)) - retail['Discount $']

# Filter mismatches
mismatches = retail[retail['Discount Diff'].abs() > 0.01]

# Print result
print(f"Found {len(mismatches)} mismatched discount amounts.")
print(mismatches[['Order No', 'Discount %', 'Discount $', 'Discount Diff']].head())
```

## 15. Check for mismatches in total calculation’


```python
# 1. Calculate the difference between Expected and Actual
retail['Total Diff'] = (retail['Retail Price'] * retail['Order Quantity']) - retail['Discount $'] + retail['Shipping Cost'] - retail['Total']

# 2. Find mismatches larger than 1 cent
mismatches = retail[retail['Total Diff'].abs() > 0.01]

# 3. Print the result
print(f"15. Found {len(mismatches)} mismatched total calculations.")
if not mismatches.empty:
    cols = ['Order No', 'Total', 'Total Diff']
    print(mismatches[cols].head())
```
