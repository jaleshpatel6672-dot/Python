## Assessment 2

```python
df1 = pd.read_csv("C:/Users/jales/Downloads/historical_automobile_sales.csv")
df1
```

## Q 1: Develop a Line chart using the functionality of pandas to show how automobile sales fluctuate from year to year.


```python
yearly_sales = df1.groupby('Year')['Automobile_Sales'].sum().reset_index()

yearly_sales.plot(kind='line', marker='o')

plt.title("Year-wise Automobile Sales Trend")
plt.xlabel("Year")
plt.ylabel("Total Sales")
plt.grid(True)

plt.show()
```

## Q 2: Plot different lines for categories of vehicle type and analyze the trend to answer the question Is there a noticeable difference in sales trends between different vehicle types during recession periods?



```python
grouped = df1.groupby(['Year', 'Vehicle_Type'])['Automobile_Sales'].sum().reset_index()

# Pivot for multiple lines
pivot_df1 = grouped.pivot(index='Year', columns='Vehicle_Type', values='Automobile_Sales')

# Plot
pivot_df1.plot(marker='o')

plt.title("Vehicle Type-wise Sales Trend")
plt.xlabel("Year")
plt.ylabel("Sales")
plt.legend(title="Vehicle Type")
plt.grid(True)
```

## Q 3: Use the functionality of Seaborn Library to create a visualization to compare the sales trend per vehicle type for a recession period with a non- recession period.



```python
grouped = df1.groupby(['Vehicle_Type', 'Recession'])['Automobile_Sales'].mean().reset_index()

plt.figure()
sns.barplot(data=grouped, x='Vehicle_Type', y='Automobile_Sales', hue='Recession')

plt.title("Sales Comparison: Recession vs Non-Recession")
plt.xlabel("Vehicle Type")
plt.ylabel("Average Sales")
plt.legend(title="Recession (0=No, 1=Yes)")

plt.xticks(rotation=45)
plt.show()
```

## Q 4: Now you want to compare the sales of different vehicle types during a recession and a non-recession period


```python
grouped = df1.groupby(['Vehicle_Type', 'Recession'])['Automobile_Sales'].mean().reset_index()

grouped['Recession'] = grouped['Recession'].map({0: 'Non-Recession', 1: 'Recession'})

plt.figure()
sns.barplot(data=grouped, x='Vehicle_Type', y='Automobile_Sales', hue='Recession')

plt.title("Vehicle Type Sales: Recession vs Non-Recession")
plt.xlabel("Vehicle Type")
plt.ylabel("Average Sales")

plt.xticks(rotation=45)
plt.show()
```
