## 31. Write a Python program to count the number of strings where the string length is 2 or more and the first and last character are same from a given list of strings.

```python
def count_strings(strings):
    count = 0
    for s in strings:
        if len(s) >= 2 and s[0] == s[-1]:
            count += 1
    return count

# Example
list1 = ['abc', 'xyz', 'aba', '1221']
print(count_strings(list1))
```

## 41. Write a Python program to check whether a list contains a sub list

```python
def is_sublist(main_list, sub_list):
    n = len(sub_list)
    for i in range(len(main_list) - n + 1):
        if main_list[i:i+n] == sub_list:
            return True
    return False

# Example
main_list = [1, 2, 3, 4, 5]
sub_list = [3, 4]

print(is_sublist(main_list, sub_list))
```

## 61. Write a Python function to calculate the factorial of a number (a nonnegative integer)

```python
def factorial(n):
    if n < 0:
        return "Factorial not defined for negative numbers"
    elif n == 0 or n == 1:
        return 1
    else:
        return n * factorial(n - 1)

# Example
num = int(input("Enter a number: "))
print("Factorial:", factorial(num))
```

## 63. Write a Python function to check whether a number is perfect or not.

```python
def is_perfect(n):
    if n <= 1:
        return False

    total = 0
    for i in range(1, n):
        if n % i == 0:
            total += i

    return total == n
```

```python
is_perfect(8)
```

## 90. Write python program that user to enter only odd numbers, else will raise an exception.

```python
# Custom exception class for even numbers
class EvenNumberError(Exception):
    pass

try:
    # Ask user input
    num = int(input("Enter an odd number: "))

    # Check if number is odd
    if num % 2 == 0:
        # Raise custom exception if number is even
        raise EvenNumberError("Error: You entered an even number!")

    print("Thank you! You entered an odd number:", num)

except ValueError:
    print("Error: Invalid input! Please enter an integer.")

except EvenNumberError as e:
    print(e)

finally:
    print("Program execution finished.")
```
